```python
import pandas as pd
import numpy as np

from sklearn.feature_extraction import text
from sklearn.metrics.pairwise import cosine_similarity

data = pd.read_csv("netflixData.csv")
data.head()
```

```python
print(data.isnull().sum())
```

```python
data = data[["Title", "Description", "Content Type", "Genres"]]
data.head()
```

```python
data = data.dropna()
```

```python
import nltk
import re
nltk.download('stopwords')
stemmer = nltk.SnowballStemmer("english")
from nltk.corpus import stopwords
import string
stopword=set(stopwords.words('english'))
```

```python
def clean(text):
    text = str(text).lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    text = [word for word in text.split(' ') if word not in stopword]
    text=" ".join(text)
    text = [stemmer.stem(word) for word in text.split(' ')]
    text=" ".join(text)
    return text
```

```python
data["Description Clean"]=data["Description"].apply(clean)
data[["Description","Description Clean"]].sample(10)
```

```python
data["Title Clean"] = data["Title"].apply(clean)
data[["Title","Title Clean"]].sample(10)
```

```python
data["Genres"]=data["Genres"].str.replace(' ','_')
data["Genres"]=data["Genres"].str.replace(',_',', ')
data["Genres"].unique()
```
